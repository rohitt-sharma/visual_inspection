```python
import cv2
import numpy as np
```

```python
def empty(a):
    pass
```

```python
# for reading image
path = "original/20220227_200350.jpg"

original_img = cv2.imread(path)
```

```python
scale_percent = 20 # percent of original size

width = int(original_img.shape[1] * scale_percent / 100)
height = int(original_img.shape[0] * scale_percent / 100)


dim = (width, height)


# resize image
img = cv2.resize(original_img, dim, interpolation = cv2.INTER_AREA)
```

```python
#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#TO INTRODUCE TRACKBARS
cv2.namedWindow("Trackbars")
cv2.resizeWindow("Trackbars", 640, 240)
cv2.createTrackbar("Hue Min", "Trackbars", 0, 359, empty) # value in open cv is from 0 to 179
cv2.createTrackbar("Hue Max", "Trackbars", 13, 359, empty)
cv2.createTrackbar("Sat Min", "Trackbars", 32, 255, empty)
cv2.createTrackbar("Sat Max", "Trackbars", 115, 255, empty)
cv2.createTrackbar("Val Min", "Trackbars", 58, 255, empty)
cv2.createTrackbar("Val Max", "Trackbars", 255, 255, empty)
```

```python
while True:
    #imgHSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)   # we convert image to hsv space
    imgHSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)   # we convert image to hsv space
    h_min = cv2.getTrackbarPos("Hue Min", "Trackbars")
    h_max = cv2.getTrackbarPos("Hue Max", "Trackbars")
    s_min = cv2.getTrackbarPos("Sat Min", "Trackbars")
    s_max = cv2.getTrackbarPos("Sat Max", "Trackbars")
    v_min = cv2.getTrackbarPos("Val Min", "Trackbars")
    v_max = cv2.getTrackbarPos("Val Max", "Trackbars")
    print(h_min,h_max,s_min,s_max,v_min,v_max)
    
    # this is the main code
    lower = np.array([h_min,s_min,v_min])
    upper = np.array([h_max, s_max, v_max])
    mask = cv2.inRange(imgHSV, lower, upper)
    #make the mask coloured it will check 2 img and if pixels match store result in variable
    # the variable below will show color wherever we have white pixel
    imgResult = cv2.bitwise_and(img,img,mask = mask)
    Gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    
    #The main code ends here
    
    cv2.imshow("Original",img)
    #cv2.imshow("hsv",imgHSV)
    cv2.imshow("mask",mask)
    cv2.imshow("img Result", imgResult)
    #cv2.imshow("img Gray", Gray)
    cv2.waitKey(1)
    
```
